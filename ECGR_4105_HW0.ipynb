```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv (r'C:\Users\homer\OneDrive\Documents\School Folder\D3.csv')
print (df)
```

```python
x1 = df.values[:,0]
x2 = df.values[:,1]
x3 = df.values[:,2]
Y = df.values[:,3]
m = len(Y)
```

```python
plt.scatter(x1,Y, color='red',marker= '+')
plt.grid() 
plt.rcParams["figure.figsize"] = (10,6) 
plt.xlabel('X1') 
plt.ylabel('Y') 
plt.title('Scatter plot of training data for X1') 
```

```python
plt.scatter(x2,Y, color='red',marker= '+')
plt.grid() 
plt.rcParams["figure.figsize"] = (10,6) 
plt.xlabel('X2') 
plt.ylabel('Y') 
plt.title('Scatter plot of training data for X2') 
```

```python
plt.scatter(x3,Y, color='red',marker= '+')
plt.grid() 
plt.rcParams["figure.figsize"] = (10,6) 
plt.xlabel('X3') 
plt.ylabel('Y') 
plt.title('Scatter plot of training data for X3') 
```

```python
X_0 = np.ones((m, 1)) 
```

```python
X_1 = x1.reshape(m, 1) 
X_2 = x2.reshape(m, 1) 
X_3 = x3.reshape(m, 1) 
```

```python
X1 = np.hstack((X_0, X_1)) 
X2 = np.hstack((X_0, X_2)) 
X3 = np.hstack((X_0, X_3)) 
```

```python
theta = np.zeros(2)
```

```python
def compute_cost(X1, Y, theta):
    predictions = X1.dot(theta) 
    errors = np.subtract(predictions, Y) 
    sqrErrors = np.square(errors) 
    J = 1 / (2 * m) * np.sum(sqrErrors) 
    return J 
```

```python
cost = compute_cost(X1, Y, theta) 
print('The cost for given values of theta_0 and theta_1 =', cost)
```

```python
def compute_cost(X2, Y, theta):
    predictions = X2.dot(theta) 
    errors = np.subtract(predictions, Y) 
    sqrErrors = np.square(errors) 
    J = 1 / (2 * m) * np.sum(sqrErrors) 
    return J 
```

```python
cost = compute_cost(X2, Y, theta) 
print('The cost for given values of theta_0 and theta_1 =', cost)
```

```python
def compute_cost(X3, Y, theta):
    predictions = X3.dot(theta) 
    errors = np.subtract(predictions, Y) 
    sqrErrors = np.square(errors) 
    J = 1 / (2 * m) * np.sum(sqrErrors) 
    return J 
```

```python
cost = compute_cost(X3, Y, theta) 
print('The cost for given values of theta_0 and theta_1 =', cost)
```

```python
def gradient_descent(X1, Y, theta, alpha, iterations):
    cost_history = np.zeros(iterations)
    for i in range(iterations): 
        predictions = X1.dot(theta) 
        errors = np.subtract(predictions, Y) 
        sum_delta = (alpha / m) * X1.transpose().dot(errors); 
        theta = theta - sum_delta; 
        cost_history[i] = compute_cost(X1, Y, theta)   
 
    return theta, cost_history 
```

```python
def gradient_descent(X2, Y, theta, alpha, iterations):
    cost_history = np.zeros(iterations)
    for i in range(iterations): 
        predictions = X2.dot(theta) 
        errors = np.subtract(predictions, Y) 
        sum_delta = (alpha / m) * X2.transpose().dot(errors); 
        theta = theta - sum_delta; 
        cost_history[i] = compute_cost(X2, Y, theta)   
 
    return theta, cost_history 
```

```python
def gradient_descent(X3, Y, theta, alpha, iterations):
    cost_history = np.zeros(iterations)
    for i in range(iterations): 
        predictions = X3.dot(theta) 
        errors = np.subtract(predictions, Y) 
        sum_delta = (alpha / m) * X3.transpose().dot(errors); 
        theta = theta - sum_delta; 
        cost_history[i] = compute_cost(X3, Y, theta)   
 
    return theta, cost_history 
```

```python
theta = [0., 0.] 
iterations = 1500; 
alpha = 0.01; 
```

```python
theta, cost_history = gradient_descent(X1, Y, theta, alpha, iterations) 
print('Final value of theta =', theta) 
print('cost_history =', cost_history) 
```

```python
plt.scatter(X1[:,1], Y, color='red', marker= '+', label= 'Training Data') 
plt.plot(X1[:,1],X1.dot(theta), color='green', label='Linear Regression') 
 
plt.rcParams["figure.figsize"] = (10,6) 
plt.grid() 
plt.xlabel('X1') 
plt.ylabel('Y') 
plt.title('Linear Regression Fit X1 w/ alpha = 0.01') 
plt.legend() 
```

```python
plt.plot(range(1, iterations + 1),cost_history, color='blue') 
plt.rcParams["figure.figsize"] = (10,6) 
plt.grid() 
plt.xlabel('Number of iterations') 
plt.ylabel('Cost (J)') 
plt.title('Convergence of gradient descent X1 w/ alpha = 0.01') 
```
